```
from newspaper import Article
import pandas as pd
# link = 'https://pt.wikipedia.org/wiki/Ci%C3%AAncia_de_dados'
link = 'https://pt.wikipedia.org/wiki/Homo_Deus:_Uma_Breve_Hist%C3%B3ria_do_Amanh%C3%A3'
article = Article(link, 'pt-br')
article.download()
article.parse()

from wordcloud import WordCloud, STOPWORDS

STOPWORDS = ['ver','essa','vez','nas','mas', 'com',
             'qual','ele','ter','pois','este',
             'vez','ver principal','artigo principal','já',
             'aos','pode','outro','artigo','desse',
             'alguns','meio','entre','das','podem','esse',
             'seu','também','são','quando','de', 'que','em',
             'os','as','da','como','dos','ou','se','um','uma',
             'para','na','ao','mais','por','não','ainda','muito','sua',
             'a', 'é', 'o', 'e', 'no', 'do', 'ser', 'nos', 'pela']

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
# dataset = open("sampleWords.txt", "r").read()
def create_word_cloud(string):
  # imageName = 'cerebro.jpg'
  imageName = 'cloud.jpeg'
  imageURL = "/content/drive/MyDrive/Colab Notebooks/" + imageName
  maskArray = np.array(Image.open(imageURL))
  cloud = WordCloud(background_color = "white", max_font_size=50, max_words = 200, mask = maskArray, stopwords = set(STOPWORDS))
  cloud.generate(string)
  cloud.to_file("wordCloud.png")
  plt.figure()
  plt.imshow(cloud, interpolation='bilinear')
  plt.axis('off')
# dataset = dataset.lower()
create_word_cloud(article.text)
```
